```python
import pandas as pd
import matplotlib.pyplot as plt

# load csv file
data = pd.read_csv('outputfile.csv', skiprows=1, usecols=[1, 5, 7])

# rename cols
data.columns = ['Number of Variables', 'Sat/Unsat', 'Execution Time']

# separate sat/unsat data
sat_data = data[data['Sat/Unsat'] == 'S']
unsat_data = data[data['Sat/Unsat'] == 'U']

# plot scatter
plt.figure(figsize=(10, 6))

# plot sat in green
plt.scatter(sat_data['Number of Variables'], sat_data['Execution Time'], marker='x', color='green', alpha=0.6, label='Sat')

# plot unsat in red
plt.scatter(unsat_data['Number of Variables'], unsat_data['Execution Time'], marker='.', color='red', alpha=0.6, label='Unsat')

# add titles
plt.title('Number of Variables vs Execution Time')
plt.xlabel('Number of Variables')
plt.ylabel('Execution Time')

# legend to show sat/unsat key data
plt.legend()

# show plot
plt.show()
```
